# 6. Часто для качественного анализа аудитории использую подходы, основанные на сегментации. Используя python, построй RFM-сегментацию пользователей, чтобы качественно оценить свою аудиторию. В кластеризации можешь выбрать следующие метрики: R - время от последней покупки пользователя до текущей даты, F - суммарное количество покупок у пользователя за всё время, M - сумма покупок за всё время. Подробно опиши, как ты создавал кластеры. Для каждого RFM-сегмента построй границы метрик recency, frequency и monetary для интерпретации этих кластеров. Пример такого описания: RFM-сегмент 132 (recency=1, frequency=3, monetary=2) имеет границы метрик recency от 130 до 500 дней, frequency от 2 до 5 заказов в неделю, monetary от 1780 до 3560 рублей в неделю. Описание подхода можно найти тут. (35 баллов)

```python
# это жесть

# будем использовать собирать все три дф в один
# order_approved_at - дата оплаты, значит продажа состоялсь
```

```python
df_customers_and_orders = df_customers.merge(df_orders, how='inner', on='customer_id')
df_customers_and_orders_and_items = df_order_items.merge(df_customers_and_orders, how='inner', on='order_id')
```

```python
df_customers_and_orders_and_items.head()
```

```python
customers_orders_6 = df_customers_and_orders_and_items
customers_orders_6.shape
#создадим датасет для 6 задачи
```

```python
customers_orders_6.head()
```

```python
customers_orders_6.order_approved_at = pd.to_datetime(customers_orders_6.order_approved_at)
```

```python
last_day = customers_orders_6.order_approved_at.max()
last_day
```

```python
from datetime import timedelta
```

```python
# на 1 день уходим вперед от последней даты
today = last_day + timedelta(days=1)
today
```

```python
# Создаём таблицу RFM

# R - время от последней покупки пользователя до текущей даты 
# F - суммарное количество покупок у пользователя за всё время 
# M - стоимость покупок за всё время
```

```python
RFM = customers_orders_6 \
    .groupby('customer_unique_id') \
    .agg({'order_approved_at': lambda x: (today - x.max()).days, 
          'product_id': lambda x: len(x), 
          'price': lambda x: x.sum()}) \
    .rename(columns={'order_approved_at': 'R', 'product_id': 'F', 'price': 'M'})
```

```python
RFM
```

```python
# Разбиваем на 5 диапазонов: до 20, 20-40, 40-60, 60-80 и выше 80 %
RFM_seg = RFM.quantile(q=[0.20, 0.40, 0.60, 0.80]).to_dict()
RFM_seg
```

```python
# функиция, которая сегментирует покупателей

def to_seg (a,b,c):
    if a <= c[b][0.20]:
        return 1
    elif a <= c[b][0.40]:
        return 2
    elif a <= c[b][0.60]:
        return 3
    elif a <= c[b][0.80]:
        return 4
    else:
        return 5
```

```python
# Применяем функцию для каждого кластера

RFM['R_seg'] = RFM['R'].apply(to_seg, args=('R', RFM_seg, ))
RFM['F_seg'] = RFM['F'].apply(to_seg, args=('F', RFM_seg, ))
RFM['M_seg'] = RFM['M'].apply(to_seg, args=('M', RFM_seg, ))
RFM
```

```python
# Склеиваем итог в одно поле

RFM['RFMseg'] = RFM.R_seg.map(str) + RFM.F_seg.map(str) + RFM.M_seg.map(str)
RFM
```
